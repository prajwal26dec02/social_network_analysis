```python
# find country given address

def findCountry(address):
  from geopy.geocoders import Nominatim
  geolocator = Nominatim(user_agent="test")
  location = geolocator.geocode(address, language="en")
  if location is None:
    return None
  return location[0].split(', ')[-1]
# findCountry("Macau, China")
```

```python
size = len(df)
for i in range(0, size):
  destination = df.iloc[i]['country']
  src = df.iloc[i]['travel_history_location'].split(';')
  # print(src)
  for address in src:
    try:
      source = findCountry(address)
      if source is None:
        continue
      covid_df.loc[len(covid_df.index)] = [source, destination] 
    except Exception as e:
      print(e, "trying for: ", address, "in" , i + 1)
  print("progress", i,"completed")
```
